```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
# Randomized Search CV
from sklearn.model_selection import RandomizedSearchCV
n_estimators=[100,200,300,400,500,800,1000,1200]
max_depth=[int(x) for x in np.linspace(10,200,10)]
#max_depth = int(max_depth)
min_samples_split=[2,5,10,15,20,25]
min_samples_leaf=[1,2,4,6,8,10,20,40,50]
max_features=['sqrt','log2','auto']
```

```python
random_grid = {
    'n_estimators':n_estimators,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf,
    'max_features':max_features
}
```

```python
rf = RandomForestRegressor(verbose=2)
h_tuned = RandomizedSearchCV(estimator=rf,param_distributions=random_grid,n_jobs=-1,cv=5)
```

```python
h_tuned.fit(x_train,y_train)
```

```python
h_tuned.best_estimator_
```

```python
h_tuned.best_params_
```

```python
rf = RandomForestRegressor(max_depth=73, max_features='sqrt', min_samples_leaf=2,
                      n_estimators=800
                      )
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
y_pred = rf.predict(x_test)
```

```python
sns.displot(y_test-y_pred,bins = 80,kde = 1)
```
