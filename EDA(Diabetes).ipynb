```python
positive = df[df['Outcome'] == 1]
negetive = df[df['Outcome'] == 0]
```

```python
positive  = positive.drop('Outcome',axis = 1)
negetive  = negetive.drop('Outcome',axis = 1)
```

```python
i = 0
plt.figure(figsize=(20, 20))
#plt.subplot(8,2,1)
columns = df.drop('Outcome',axis = 1).columns
for col in columns:
  plt.subplot(4,2,i+1)
  plt.hist(positive[col],bins=60,alpha = 0.5,label = 'Positive')
  
  plt.hist(negetive[col],bins = 60,alpha = 0.2,color = 'red',label='Negetive')
 # plt.hist(negetive[col],color = 'red')
  plt.legend()
  plt.xlabel(columns[i])
  plt.ylabel('Count')
  i = i + 1 
```
