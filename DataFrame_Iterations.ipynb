```python
#Class Creation within a dataset
# Ex: AGE = 30 = ADULT
#            15 = TEENEGER
# For this wee use .apply() method
def convert_class(x):
  if x<10:
    return 'child'
  if x>10 and x<45:
    return 'adult'
  else:
    return 'old'
df['Age'] = df['Age'].apply(convert_class)
```
