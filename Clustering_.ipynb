```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=5)
```

```python
model = model.fit(x,y)
```

```python
centres = model.cluster_centers_
centres
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(centres[:,0],centres[:,1],color = 'red',label='Centre',marker='*')
plt.legend()
plt.show()
```

```python
ypred = model.predict(x)
```

# KMeans vs DBScan

```python
from sklearn.datasets import make_circles,make_moons
```

```python
x,y = make_moons(n_samples=200,noise=0.1)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
model = KMeans(n_clusters=2)
model = model.fit(x,y)
```

```python
centres = model.cluster_centers_
centres
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(centres[:,0],centres[:,1],color = 'red',label='Centre',marker='*')
plt.legend()
plt.show()
## See it's not able to do a good quality clustering.
```

# DBScan
- Density Based Spatial Clustering Applications with noise.
- Based upon the idea, that a cluster is a high density area surrounded by low density region.
- Stars by exploring the small area, if density is "good enough" it considered as a part of cluster, and explores neighbours and so on, so it groups regions based upon density.

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(min_samples=5,eps = 0.2)
```

```python
model=model.fit(x,y)
```

```python
ypred = model.fit_predict(x)
ypred
```

```python
plt.scatter(x[:,0],x[:,1],c=ypred)
plt.show()
```
