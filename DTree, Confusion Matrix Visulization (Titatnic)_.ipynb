Dummy Variables

```python
dummies = pd.get_dummies(merged['Sex'],drop_first=True)
merged = pd.concat([merged,dummies],axis = 'columns')
dummies = pd.get_dummies(merged['Embarked'],drop_first=True)
merged = pd.concat([merged,dummies],axis = 'columns')
merged.head()
```
